```python
color_rules = {
    ("blue", "bright"): ["white", "cream", "light_yellow", "nude", "metallic"],
    ("blue", "navy"): ["white", "cream", "brown", "metallic"],
    ("yellow", "bright"): ["white", "black", "blue"],
    ("black", "dark"): ["white", "red", "cream", "metallic"]
}
```

```python
pattern_rules = {
    "plain": ["plain", "floral", "stripes", "polka_dot"],
    "floral": ["plain"],
    "stripes": ["plain"],
    "animal": ["plain"],
    "ethnic": ["plain"],
    "geometric": ["plain"],
    "cartoon": ["plain"]
}
```

```python
occasion_rules = {
    "party": ["party"],
    "casual": ["casual", "smart_casual"],
    "office": ["office", "smart_casual"]
}
```

```python
season_rules = {
    "summer": ["summer", "spring"],
    "winter": ["winter", "autumn"]
}
```

```python
fabric_rules = {
    "cotton": ["cotton", "linen"],
    "linen": ["linen", "cotton"],
    "silk": ["silk", "satin"],
    "denim": ["denim"],
    "wool": ["wool", "knit"]
}
```

```python
body_type_rules = {
    "pear": {
        "fit": ["wide_leg", "straight"],
        "neckline": ["v_neck", "boat_neck"]
    },

    "apple": {
        "fit": ["bootcut"],
        "neckline": ["v_neck"]
    }
}
```

```python
skin_tone_rules = {
    "fair": ["blue", "emerald", "red", "black"],
    "warm": ["olive", "brown", "mustard"],
    "deep": ["white", "orange", "gold"]
}
```

```python
sleeve_rules = {
    "full_sleeve": ["winter", "autumn"],
    "short_sleeve": ["summer", "spring"],
    "sleeveless": ["summer"]
}
```

```python
def recommend(user_item, products):
    results = []

    for product in products:
        score = 0
        reasons = []

        # category
        if user_item["category"] == "tops" and product["category"] == "bottoms":
            score += 15
            reasons.append("Top pairs well with bottom")

        # color + tone
        allowed_colors = color_rules.get(
            (user_item["color"], user_item["tone"]),
            []
        )

        if product["color"] in allowed_colors:
            score += 20
            reasons.append("Color shade matches well")

        # pattern
        allowed_patterns = pattern_rules.get(user_item["pattern"], [])

        if product["pattern"] in allowed_patterns:
            score += 15
            reasons.append("Pattern combination is balanced")

        # occasion
        allowed_occasions = occasion_rules.get(user_item["occasion"], [])

        if any(occ in product["occasion"] for occ in allowed_occasions):
            score += 15
            reasons.append("Suitable for the occasion")

        # season
        allowed_seasons = season_rules.get(user_item["season"], [])

        if any(season in product["season"] for season in allowed_seasons):
            score += 10
            reasons.append("Suitable for this season")

        # fabric
        allowed_fabrics = fabric_rules.get(user_item["fabric"], [])

        if product["fabric"] in allowed_fabrics:
            score += 10
            reasons.append("Fabric matches well")

        # body type fit
        body_rules = body_type_rules.get(user_item["body_type"], {})

        if product["fit"] in body_rules.get("fit", []):
            score += 10
            reasons.append("Fit suits your body type")

        # formality
        if product["formality"] == user_item["formality"]:
            score += 5
            reasons.append("Matches your formality preference")

        results.append({
            "product": product,
            "score": min(score, 100),
            "reasons": reasons
        })

    return sorted(results, key=lambda x: x["score"], reverse=True)
```

```python
def get_match_label(score):
    if score >= 85:
        return "Excellent Match"
    elif score >= 70:
        return "Good Match"
    elif score >= 50:
        return "Okay Match"
    else:
        return "Weak Match"
```

```python
def show_recommendations(results, top_n=3):
    for i, item in enumerate(results[:top_n], start=1):
        product = item["product"]
        label = get_match_label(item["score"])

        print(f"{i}. {product['name']}")
        print(f"   Match Score: {item['score']}/100 - {label}")
        print(f"   Color: {product['color']}")
        print(f"   Pattern: {product['pattern']}")
        print(f"   Fabric: {product['fabric']}")
        print("   Why:")
        for reason in item["reasons"]:
            print("   -", reason)
        print()
```

```python
results = recommend(user_item, products)
show_recommendations(results, top_n=3)
```
